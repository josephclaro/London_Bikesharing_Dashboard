```python
#london bikeshare dataset
import pandas as pd

bikes = pd.read_csv("london_merged.csv")
bikes

# cnt: count of new bikeshares per hour
# t1: real temperature
# t2: "feels-like" temperature
# hum: humidity (%)
# wind_speed (km/h)
# is_holiday: 1-holiday, 0-regular day
# is_weekend: 1-weekend, 0-weekday
# season: 0-spring, 1-summer, 2-autumn, 3-winter
# weather-code: weather categories
#   1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity
#   2 = scattered clouds / few clouds
#   3 = Broken clouds
#   4 = Cloudy
#   7 = Rain/ light Rain shower/ Light rain
#   10 = rain with thunderstorm
#   26 = snowfall
#   94 = Freezing Fog
```

```python
# counts of each weather category
bikes.weather_code.value_counts()
```

```python
# Specify new column names as dictionary
newBikeCols = {
    'timestamp': 'date_time',
    'cnt': 'count',
    't1': 'real_temp_C',
    't2': 'feels_like_temp_C',
    'hum': 'humidity',
    'wind_speed': 'wind_speed_kmh',
    'weather_code': 'weather',
    'is_holiday': 'is_holiday',
    'is_weekend': 'is_weekend',
    'season': 'season'
}

# Rename dataFrame columns
bikes.rename(newBikeCols, axis=1, inplace = True)
```

```python
# Change humidity values to between 0,1
bikes.humidity = bikes.humidity / 100
```

```python
# Season dictionary (call seasons by name)
season_dict = {
    '0.0': 'spring',
    '1.0': 'summer',
    '2.0': 'autumn',
    '3.0': 'winter'
}

# Weather dictionary
weather_dict = {
    '1.0': 'clear',
    '2.0': 'scattered cloud',
    '3.0': 'moderate cloud',
    '4:0': 'cloudy',
    '7.0': 'rain',
    '10.0': 'rain with thumderstorm',
    '26.0': 'snowfall'
}

# Change column data types from int to string
bikes.season = bikes.season.astype('str')
bikes.weather = bikes.weather.astype('str')

# Update values in columns to string versions
bikes.season = bikes.season.map(season_dict)
bikes.weather = bikes.weather.map(weather_dict)
bikes.head()
```

```python
bikes.to_csv('london_bikes_final.csv')
```
